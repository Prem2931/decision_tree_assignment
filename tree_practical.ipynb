Qn16>>Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy.

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.DESCR)

```

```python
X_iris, y_iris = iris.data, iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.2, random_state=1)
```

```python
from sklearn.tree import  DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train_iris, y_train_iris)
```

```python
ypred_iris = dt.predict(X_test_iris)
ypred_iris
```

```python
from sklearn.metrics import  accuracy_score
print("Accuracy:", accuracy_score(y_test_iris, ypred_iris))
```

Qn19>>Write a Python program to train a Decision Tree Regressor on a housing dataset and evaluate using Mean
Squared Error (MSE).

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
X_housing, y_housing = housing.data, housing.target
```

```python
X_train_housing, X_test_housing, y_train_housing, y_test_housing = train_test_split(X_housing, y_housing, test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(X_train_housing, y_train_housing)
y_pred_housing = dt_regressor.predict(X_test_housing)
```

```python
from sklearn.metrics import  mean_squared_error
print("MSE: ", mean_squared_error(y_test_housing, y_pred_housing))
```

Qn23>>Write a Python program to apply feature scaling before training a Decision Tree Classifier and compare its
accuracy with unscaled data

```python
from sklearn.preprocessing import  StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_iris)
X_test_scaled = scaler.transform(X_test_iris)
dt_scaled = DecisionTreeClassifier()
dt_scaled.fit(X_train_scaled, y_train_iris)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
print("Accuracy (Scaled Data):", accuracy_score(y_test_iris, y_pred_scaled))
print("Accuracy (Unscaled Data):", accuracy_score(y_test_iris, ypred_iris))
```

Qn24>>Write a Python program to train a Decision Tree Classifier using One-vs-Rest (OvR) strategy for multiclass

```python
from sklearn.multiclass import OneVsRestClassifier
dt_ovr = OneVsRestClassifier(DecisionTreeClassifier())
dt_ovr.fit(X_train_iris, y_train_iris)
y_pred_ovr = dt_ovr.predict(X_test_iris)
print("Accuracy (One-vs-Rest):", accuracy_score(y_test_iris, y_pred_ovr))
```
