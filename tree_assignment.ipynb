Qn1>What is a Decision Tree, and how does it work?

Ans>

A Decision Tree is a supervised learning algorithm used for classification and regression. It works by splitting the dataset into subsets based on feature values, forming a tree structure where each internal node represents a decision rule, each branch represents an outcome, and each leaf node represents a class label (for classification) or a value (for regression).

Qn2>>What are impurity measures in Decision Trees?

Ans>>
Impurity measures help determine how well a split separates the data. Common impurity measures include:

Gini Impurity (used in CART)

Entropy (used in ID3 and C4.5)

Variance reduction (for regression trees)

Qn6>>What is the difference between Gini Impurity and Entropy?

Ans>>

Gini Impurity measures how often a randomly chosen element would be incorrectly classified if randomly labeled.

Entropy measures the amount of uncertainty in the data.

Gini is computationally faster, while Entropy gives more weight to less frequent classes.

Qn7>>What is the mathematical explanation behind Decision Trees?

Ans>>

Decision Trees use recursive partitioning to minimize impurity measures (Gini or Entropy). The tree grows by selecting features that provide the highest Information Gain and stops when further splits do not improve classification.


Qn8>>What is Pre-Pruning in Decision Trees?
Pre-Pruning stops tree growth early based on predefined conditions like:

Maximum tree depth

Minimum number of samples per node

Qn10>>What is the difference between Pre-Pruning and Post-Pruning?

Ans>>

Pre-Pruning stops tree growth early, preventing overfitting but possibly leading to underfitting.

Post-Pruning builds the full tree first and then simplifies it, usually providing better generalization.

Qn11>>What is a Decision Tree Regressor?
Ans>>

A Decision Tree Regressor predicts continuous values instead of class labels. It uses mean squared error (MSE) or variance reduction as the impurity measure to find the best splits.

What are the advantages and disadvantages of Decision Trees?

Ans>>

Advantages:

Easy to interpret and visualize

Handles numerical and categorical data

Requires little data preprocessing

Disadvantages:

Prone to overfitting

Biased towards dominant classes if data is imbalanced


Can be unstable (small changes in data can lead to different trees)

Qn13>>How does a Decision Tree handle missing values?

Ans>>

By ignoring missing values during splitting
By imputing missing values (e.g., replacing with mean, median, or mode)
Using surrogate splits (alternative splitting criteria)

Qn14>>How does a Decision Tree handle categorical features?

Ans>>

By one-hot encoding (for algorithms that require numerical data)

By splitting based on category grouping (e.g., treating each category as a separate branch)

Qn15>>What are some real-world applications of Decision Trees?

Ans>>

Medical diagnosis (identifying diseases based on symptoms)

Customer segmentation (targeting marketing strategies)

Fraud detection (identifying fraudulent transactions)

Credit risk assessment (predicting loan defaults)

Recommendation systems (suggesting products or content)
